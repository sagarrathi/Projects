* Match winnign Percentage

```python
t_name="Sunrisers Hyderabad"
team_dict={"name":t_name}
```

```python
def team_df(t_name, mf):
    mft=mf[ (mf["team1"]==t_name) | (mf["team2"]==t_name)]
    return mft
mft=team_df(t_name, mf)
```

#  Dictonary Updater

```python
def update_dict(team_dict, variable_name, value):
    team_dict[variable_name]=value
```

#  1. Toss winning Percentage

```python
def toss_winning_percentage(mft, t_name, team_dict):
    num_of_matches=mft.count()["id"]
    won_for=mft[mft["toss_winner"]==t_name].count()["id"]
    winning_percentage=100*won_for/num_of_matches
    winning_percentage=round(winning_percentage, 2)
    update_dict(team_dict, "toss_winning_percenatge", winning_percentage)
toss_winning_percentage(mft, t_name, team_dict)
team_dict
```

#  2. Match winning Percentage

```python
def match_winning_percentage(mft, t_name,  team_dict):
    num_of_matches=mft.count()["id"]
    won_for=mft[mft["winner"]==t_name].count()["id"]
    winning_percentage=100*won_for/num_of_matches
    winning_percentage=round(winning_percentage, 2)
    update_dict(team_dict, "match_winning_percenatge", winning_percentage)
match_winning_percentage(mft,t_name,  team_dict)
team_dict
len(mft)
```

# 3. Favourite style

```python
def favourite_style(mft,t_name,  team_dict):
    won_df=mft[mft["toss_winner"]==t_name]
    
    try:
        field=won_df.groupby("toss_decision").count()["id"]["field"]
    except:
        field=0
    try:
        bat=won_df.groupby("toss_decision").count()["id"]["bat"]
    except:
        bat=0
    update_dict( team_dict, "field", field)
    update_dict( team_dict, "bat", bat)
    
favourite_style(mft, t_name, team_dict)
team_dict
```

# 4. Infamous venue

```python
def infamous_venue(mf, mft, t_name,team_dict ):
    stadium_list=list(set(mf["venue"]))
    total_matches=mft.groupby("venue").count()["id"]
    total_count=[]
    won_count=[]

    for stadium in stadium_list:
        ## If you played in stadium

        try:
            total=total_matches[stadium]

            ### If You Win
            try:
                w_mft=mft[mf["winner"]==t_name]
                won_matches=w_mft.groupby("venue").count()["id"]
                won=won_matches[stadium]
            except:
                won=0



        except:
            total="NP"
            won="NP"

        total_count.append(total)
        won_count.append(won)

    loss_count_p=[]

    for total, won in zip(total_count, won_count):
        if total != "NP":
            loss_percentage=100*(1-(won/total))
            loss_percentage=round(loss_percentage,2)
        else:
            loss_percentage="NP"

        loss_count_p.append(loss_percentage)
    loss_count_p

    loss_dict=dict(zip(stadium_list, loss_count_p))
    team_dict={**team_dict, **loss_dict}
    return team_dict
    
    
team_dict=infamous_venue(mf, mft, t_name, team_dict)
team_dict
```

# 5. winning style

```python
def winnig_style(mft,t_name, team_dict):
        
    won_for=mft[mft["winner"]==t_name]
    run_win_style=list(won_for["win_by_runs"])
    wicket_win_style=list(won_for["win_by_wickets"])
    run_win_style
    wicket_win_style
    
    run_bins = [1,25,50,100,150]
    run_hist=np.histogram(run_win_style, run_bins)

    wicket_bins = [1,3,5,7,10]
    wicket_hist=np.histogram(wicket_win_style,wicket_bins)
    
    for h in range(0,4):
            key_name=str(run_hist[1][h])+"-"+str(run_hist[1][h+1])+"r"

            try: 
                value=run_hist[0][h]
            except:
                value=0

            team_dict[key_name]=value


    for h in range(0,4):
            key_name=str(wicket_hist[1][h])+"-"+str(wicket_hist[1][h+1])+"w"

            try: 
                value=wicket_hist[0][h]
            except:
                value=0

            team_dict[key_name]=value

winnig_style(mft, t_name, team_dict)
team_dict
```

# Creating Aggregator Function


```python
def complete_aggregator(mf,t_name):
    team_dict={"name":t_name}
    
    mft=team_df(t_name, mf)
    
    #1.Toss winnign Percenatge
    toss_winning_percentage(mft,t_name, team_dict)
    #2.Match winnign Percentage
    match_winning_percentage(mft,t_name, team_dict)
    #3.Favourite style
    
    favourite_style(mft, t_name, team_dict)
    #4.Infamous venue
    team_dict=infamous_venue(mf, mft, t_name, team_dict)
    #5.winning style
    winnig_style(mft, t_name, team_dict)
    
    return team_dict
```

```python
def save_all(team_list):
    team_dict={}
    
    for team in team_list:
        t_name=str(team)
        team_dict=complete_aggregator(mf,t_name)
        team_dicts.append(team_dict)

    return team_dicts
```

```python
team_list=list(set(mf["team1"]))
team_dicts=[]

team_dicts=save_all(team_list)
```

```python
mfta=pd.DataFrame(team_dicts)
mfta.to_csv("./output/teams.csv")
mfta.head()
```
