## Search Links

### Example: 


http://search.espncricinfo.com/ci/content/player/search.html?search=sachin+tendulkar&x=0&y=0

```python
def search_link(player_list):
    search_list=[]
    
    base_link="http://search.espncricinfo.com/ci/content/player/search.html?search="
    last_link="&x=0&y=0"
    
    for player in player_list:
        player_link=str(player).replace(" ", "+")
        final_link=base_link+player_link+last_link   
        search_list.append(final_link)
    
    return search_list
```

```python
search_list=search_link(list(df["Name"]))
df["Search"]=search_list
df.head()
```

# Profile retrival

```python
def profile_link(search_list):
    profile_list=[]
    for search in search_list:
        page=urllib.request.urlopen(search)
        soup=BeautifulSoup(page)
        html_code=soup.find("a", attrs={"class":"ColumnistSmry"})
        
        if html_code:
            profile_link=html_code.get("href")
        else:
            profile_link="---"
        
        
        profile_list.append(profile_link)
    
    
    return profile_list

```

```python
def image_link(df):
    base_url="http://www.espncricinfo.com"
    
    for player in df.index:
        print("Starting for:",player)
        
        if df.loc[player]["Profile"] !="---" and not df.loc[player]["Image"]:
            
            profile_url=base_url+df.loc[player]["Profile"]
        
            print("Profile:", profile_url)
            
            page=urllib.request.urlopen(profile_url)
            soup=BeautifulSoup(page)
            image_url=soup.find("link", attrs={"rel":"image_src"})    
            
            if image_url:
                image_url=image_url.get("href")
                print("Image:", image_url)
                df.at[player,"Image"]=image_url
                
            if not image_url:
                
                    print("No Image")
        else:
            print("No Profile")
                    
        
            
```

```python
from PIL import Image
import os.path
import numpy as np
```

```python
def save_image(df):
    
    for player in df.index:
        print("For Player:",player)
        =
        local_url="./Output/Player_Images/"+str(player)+".jpg"
        
        if os.path.exists(local_url):
            print("Image Present")
            
        else:    
            print("Image Link:",df.loc[player]["Image"])
            
            if not pd.isnull(df.loc[player]["Image"]):
                image_url=df.loc[player]["Image"]
                urllib.request.urlretrieve(image_url, local_url)
                print("Image Saved")
            else:
                print("Making Dummy Image")
                img = Image.new('RGB', (200, 200), color = 'red')
                img.save(local_url+".jpg")
            
```

```python
save_image(df)
```
